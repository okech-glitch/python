```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data=[
    ['Beef', 'Pork', 'Mutton'],
    ['Pork', 'Mutton', 'Chicken'], 
    ['Mutton', 'Chicken'],
    ['Beef', 'Mutton', 'Chicken'],
    ['Beef', 'Pork', 'Mutton', 'Chicken'],
]
```

```python
#convert to a format appropriate for apriori
encoder=TransactionEncoder()
encoded=encoder.fit_transform(data)
dataframe=pd.DataFrame(encoded, columns=encoder.columns_)
dataframe
```

```python
#apply apriori
frequent_items=apriori(dataframe, min_support=0.3, use_colnames=True)
frequent_items
```

```python
#create rules
rules=association_rules(frequent_items, min_threshold=0.7, metric='confidence')
rules
```
