```python
import pandas as pd
sales_dimension=pd.read_csv('sales_dimension.csv')
fact_table=pd.read_csv('fact_table.csv')
time_dimension=pd.read_csv('time_dimension.csv')
```

```python
MergedData=pd.merge(sales_dimension,fact_table, on='salesid')
MergedData
```

```python
FinalData=pd.merge(MergedData, time_dimension, on='timeid')
FinalData
```

```python
quartelysales=FinalData.groupby(['quarter'])['sales'].sum()
quartelysales
```
