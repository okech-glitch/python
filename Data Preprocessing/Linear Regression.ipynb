```python
import pandas as pd
import seaborn as sb
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn import datasets,linear_model
from sklearn.linear_model import LinearRegression

```

```python
clientspending=pd.read_csv('clientspending.csv')
print(clientspending)
```

```python
trainingset,testingset=train_test_split(clientspending, test_size=0.3, random_state=2)
print('Training data is:\n', trainingset)
print('Testing data is:\n', testingset)
```

```python
#split the training data into independent and dependent variables
independentvar=trainingset[['Monthly_Income_pounds']].values
dependentvar=trainingset[['Expenditure_pounds']].values
print('Training Independent Variable is:\n', independentvar)
print('Training Dependent Variable is:\n', dependentvar)
```

```python
#create the Linear Regression model
regression=LinearRegression()
regression.fit(independentvar, dependentvar)
print('The coefficient of the model is:\n', regression.coef_)
```

```python
#Relationship between the annual income and expenditure variables in training data
plt.scatter(independentvar, dependentvar, color='green')
plt.title('Data for training set')
plt.xlabel('Annual Income')
plt.ylabel('Expenditure')
plt.grid()
plt.show()
```

```python
#display the linear regression meodel
plt.figure(figsize=(10,6))
sb.regplot(x=independentvar.flatten(), y=dependentvar.flatten(), data=trainingset, line_kws={"color":"red"}, scatter_kws={"color":"green"})
plt.title("Client Data for the Training Set")
plt.xlabel("Monthly Income (Pounds)")
plt.ylabel("Expenditure (Pounds)")
plt.grid()
plt.show()
```
