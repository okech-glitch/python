```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X=[3,21,22,34,54,34,55,67,89,99]
Y=[1,10,14,34,44,36,22,67,79,90]
X_array=np.array(X).reshape(-1, 1)
Y_array=np.array(Y)
X_array,Y_array
```

```python
#show relationship bewtween X and Y
from matplotlib import pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(X_array, Y_array, marker='o', color='green')
plt.grid()
plt.title('Relationship between X and Y')
plt.xlabel('X_array')
plt.ylabel('Y_array')
```

```python
#apply Linear Regression and print the coefficient and R2(score)
regression=LinearRegression()
regression.fit(X_array, Y_array)
print(regression.coef_)
regression.score(X_array, Y_array)
```
