```python
import pandas as pd
workers=pd.read_csv('workers.csv')
workers
```

```python
#drop missing values
workers=workers.dropna()
workers
```

```python
workers.describe()
```

```python
# Use a relevant function to filter and display the data for all employees earning a salary greater or equal to 5000 dollars. 
GreaterSalary=workers[workers['Salary']>=5000]
GreaterSalary
```

```python
#(f) Calculate and display the total salary paid to the staff by department and education level.  (4 marks) 
SalaryPaid=workers.groupby(['Dept','EducationLevel'])['Salary'].sum()
SalaryPaid
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(10,6))
SalaryPaid.plot(kind='bar', color='skyblue')
```
