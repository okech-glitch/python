```python
import pandas as pd
supermarket=pd.read_csv('supermarket.csv')
supermarket
```

```python
#check for null values
supermarket.isna().sum()
```

```python
#fill null values with the mean
supermarket['Price'].fillna(supermarket['Price'].mean(), inplace= True)
supermarket
```

```python
#confirm
supermarket.isna().sum()
```
