```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data=[
    ['Beef', 'Pork', 'Mutton'],
    ['Pork', 'Mutton', 'Chicken'], 
    ['Mutton', 'Chicken'],
    ['Beef', 'Mutton', 'Chicken'],
    ['Beef', 'Pork', 'Mutton', 'Chicken'],
]
```

```python
#convert the data into a form suitable for apriori
encoder=TransactionEncoder()
encoded=encoder.fit_transform(data)
dataframe=pd.DataFrame(encoded, columns=encoder.columns_)
dataframe
```

```python
#apply apriori
frequentitems=apriori(dataframe, min_support=0.4, use_colnames=True)
frequentitems
```

```python
#create rules
rules=association_rules(frequentitems, metric="confidence", min_threshold=0.7)
rules
```
