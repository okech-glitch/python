```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
customer=pd.read_csv('customerdata.csv')
customer
```

```python
scaler=MinMaxScaler()
scaler.fit(customer[['AnnualIncome','SpendingScore']])
customer[['AnnualIncome','SpendingScore']]=scaler.transform(customer[['AnnualIncome','SpendingScore']])
customer[['AnnualIncome','SpendingScore']]
```

```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
X=customer[['AnnualIncome','SpendingScore']]
values=[]
for k in range(1, 10):
    kmeans=KMeans(n_clusters=k, random_state=20)
    kmeans.fit(X)
    values.append(kmeans.inertia_)
plt.figure(figsize=(8,5))
plt.plot(range(1, 10), values, marker='o')
plt.grid()
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('No. of Clusters')
plt.ylabel('Intertia')
plt.show()
```
