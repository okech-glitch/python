```python
import pandas as pd
regional_sales=pd.read_csv('regional_sales.csv')
regional_sales
```

```python
product_info=pd.read_csv('product_info.csv')
product_info
```

```python
#merging the two tables
merged_data=pd.merge(regional_sales, product_info, on='ProductID')
merged_data
```

```python
#details of products whose sales are greater than 60000
filtered_data=merged_data[merged_data['Sales']>60000]
filtered_data
```

```python
#load the filtered data to a CSV file which is not indexed
filtered_data.to_csv('warehouse.csv', index='false')
```
